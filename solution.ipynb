# AI/ML Intern Assignment - Notebook Solution

This notebook implements:
1. Chronological parsing of `conversations.csv`
2. Topic checkpoints with summaries
3. Every-100-message checkpoints
4. RAG retrieval combining summaries and chunks
5. Evidence-based persona extraction
6. Outputs for chatbot consumption

```python
# Install dependencies if needed (uncomment when running first time)
# !pip install pandas numpy scikit-learn pyarrow streamlit

from pathlib import Path
import json
import pandas as pd

from rag_persona_system import ConversationRAGSystem
```

```python
DATASET_PATH = Path('conversations.csv')
ARTIFACT_DIR = Path('artifacts')

assert DATASET_PATH.exists(), f'Missing dataset at {DATASET_PATH.resolve()}'
ARTIFACT_DIR.mkdir(parents=True, exist_ok=True)

system = ConversationRAGSystem(chunk_size=8, chunk_overlap=2)
stats = system.run_pipeline(DATASET_PATH, ARTIFACT_DIR)
stats
```

```python
# Show sample normalized messages
messages_df = system.messages_df
messages_df.head(10)
```

```python
# Topic checkpoint overview
topic_df = pd.DataFrame(system.topic_checkpoints)
print('Total topics:', len(topic_df))
topic_df.head(10)
```

```python
# Every-100-message checkpoint overview
hundred_df = pd.DataFrame(system.hundred_checkpoints)
print('Total 100-message checkpoints:', len(hundred_df))
hundred_df.head(10)
```

```python
# RAG query examples
queries = [
    'What are this user\'s habits?',
    'What kind of person is this user?',
    'How do they usually talk?',
]

rag_outputs = {}
for q in queries:
    rag_outputs[q] = system.answer_query(q, k_summaries=5, k_chunks=6)
    print('\n' + '='*80)
    print('Query:', q)
    print(rag_outputs[q]['answer'][:1200])
```

```python
# Persona JSON preview
persona = system.persona
print(json.dumps(persona, indent=2)[:4000])
```
